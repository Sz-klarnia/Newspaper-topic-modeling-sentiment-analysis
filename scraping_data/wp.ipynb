```python
import time
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
# creating file, writing headers
f = open("wp.csv","w",encoding="utf8")
headers = "Title^Author^Date^Link^Content\n"
f.write(headers)


base_url = "https://wiadomosci.wp.pl/"
page_nr = 67
month = "-02"

while "-01" not in month:
    print(page_nr)
    url = base_url + str(page_nr)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    links = soup.find_all("div", class_="g2PrHTUx")
    
    for link in links:
        # getting informations from page
        url_page = base_url[:-1] + link.find("a").get("href")
        title = link.find("a").get("title")
        title = title.strip().replace("\n","")
        
        # requesting and parsing article
        article_page = requests.get(url_page)
        while str(article_page) != "<Response [200]>":
            article_page = requests.get(url_page)
        soup_page = BeautifulSoup(article_page.text, "html.parser")        
        try:
            author = soup_page.find("span", class_="signature--author g2aU53vl desktop").text
        except:
            author= "None"
            
        try:
            date = soup_page.find("time").text
        except AttributeError:
            continue
        if "-02" not in date:
            continue

        # getting article content
        content = ""
        try:
            header = soup_page.find("div",class_="article--lead g1HGmjUl").text
        except:
            continue
        content +=header
        texts = soup_page.find_all("div",class_="article--text gFQN8OU2 aYwaUr3X")
        for p in texts:
            content += p.text + " "
        content = content.strip().replace("\n","")
        content = content.replace("\xa0"," ")
        
        f.write(title + "^" + author + "^" + date + "^" + url_page + "^" + content + "\n")
        month = date

    page_nr +=1
    
f.close()
```
