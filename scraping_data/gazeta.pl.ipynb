```python

import time
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
# creating file, writing headers
f = open("gazeta_pl.csv","w",encoding="utf8")
headers = "Title^Author^Date^Link^Content\n"
f.write(headers)

page_nr = 1
month = ".02"

while ".01" not in month:
    url = f"https://wiadomosci.gazeta.pl/wiadomosci/0,114871.html?str={page_nr}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    links = soup.find_all("li", class_="entry")
    for link in links:
        # getting informations from page
        author = link.find("span", class_="author").text.strip()
        date = link.find("time").text
        if ".01." in date:
            break
        if ".02." not in date:
            continue

        url_page = link.find("a").get("href")
        title = link.find("a").get("title")
        title = title.strip().replace("\n","")
        
        # requesting and parsing article
        article_page = requests.get(url_page)
        soup_page = BeautifulSoup(article_page.text, "html.parser")
        # getting article content
        try:
            header = soup_page.find("div",attrs={"id":"gazeta_article_lead"}).text
        except:
            continue
        content = header + " "
        contents = soup_page.find_all("p",class_="art_paragraph")
        for p in contents:
            content+= p.text + " "
        content = content.strip().replace("\n","")
        content = content.replace("\xa0"," ")
        
        f.write(title + "^" + author + "^" + date + "^" + url_page + "^" + content + "\n")
    
    page_nr +=1
    print(page_nr)
    month = date[2:5]
    
f.close()
```
