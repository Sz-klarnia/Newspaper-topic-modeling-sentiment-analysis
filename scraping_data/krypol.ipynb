```python

import time
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
from selenium import webdriver
```

```python
driver = webdriver.Firefox()

driver.get("https://krytykapolityczna.pl/kraj/")
time.sleep(2)


for i in range(10):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(2)


```

```python
# creating file, writing headers
f = open("krypol.csv","w",encoding="utf8")
headers = "Title^Author^Date^Link^Content\n"
f.write(headers)




soup = BeautifulSoup(driver.page_source, "html.parser")

links = soup.find_all("article")

for link in links:
    # getting informations from page
    author = link.find("span", itemprop="name").text
    date = link.find("time").text
    if "lutego" not in date:
        continue
    url_page = link.find("a",class_="post-preview").get("href")
    title = link.find("h2").text
    title = title.strip().replace("\n","")
    
    # requesting and parsing article
    article_page = requests.get(url_page)
    soup_page = BeautifulSoup(article_page.text, "html.parser")
    # getting article content
    content = " "
    texts = soup_page.find_all("p")
    for p in texts:
        content += p.text + " "
    content = content.strip().replace("\n","")
    content = content.replace("\xa0"," ")
    
    f.write(title + "^" + author + "^" + date + "^" + url_page + "^" + content + "\n")


    
f.close()
```

```python
data = pd.read_csv("krypol.csv",sep="^")
```
