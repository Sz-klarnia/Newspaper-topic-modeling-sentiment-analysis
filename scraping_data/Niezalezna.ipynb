```python

import time
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
# creating file, writing headers
f = open("NIezalezna_polska.csv","w",encoding="utf8")
headers = "Title^Author^Date^Link^Content\n"
f.write(headers)


base_url = "https://niezalezna.pl/dzial/polska/"
page_nr = 1
month = ".02"

while month != ".01":
    url = base_url + str(page_nr)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    links = soup.find_all("a", class_="uitemUnderline")
    
    for link in links:
        page_url = "https://niezalezna.pl" + link.attrs["href"]
        
        # requesting and parsing article
        article_page = requests.get(page_url)
        soup_page = BeautifulSoup(article_page.text, "html.parser")
            
        title = soup_page.find("h1").text
        title = title.strip().replace("\n","")

        date = soup_page.find("span",class_="newsTime").text
        if date[2:5] != ".02":
            continue
        author = soup_page.find("p",class_="authorDesc").text
        
        # getting article content
        header = soup_page.find("div",class_ = "articleLead").text
        contents = soup_page.find_all("div", class_="articleBody")
        content = header + " "
        for lines in contents:
            content += lines.text + " "

        content = content.strip().replace("\n","")
        content = content.replace("\xa0"," ")
        
        f.write(title + "^" + author + "^" + date + "^" + page_url + "^" + content + "\n")
    
    page_nr +=1
    month = date[2:5]
    
f.close()
```
