```python

import time
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Firefox()

driver.get("https://wpolityce.pl/polityka")
time.sleep(2)

elem = driver.find_element_by_xpath("/html/body/div[1]/div/div/div/div[2]/div/button[2]")
elem.click()
time.sleep(2)

for i in range(30):
    driver.execute_script("arguments[0].scrollIntoView();", driver.find_element_by_id("wPolityce/div_interstitial"))
    time.sleep(2)
```

```python
# creating file, writing headers
f = open("wPl_polityka.csv","w",encoding="utf8")
headers = "Title^Author^Date^Link^Content\n"
f.write(headers)

basic_link = "https://wpolityce.pl"
soup = BeautifulSoup(driver.page_source, "html.parser")

links = soup.find_all("div",class_="tile--basic")

for link in links:
    try:
        url = basic_link + link.find("a",class_="tile__link").get('href')
    except:
        continue
    page = requests.get(url)
    soup1 = BeautifulSoup(page.text, "html.parser")
    title = soup1.find("h1",class_="article__title").text
    date = soup1.find("time",class_="js-relative-date").text
    if "-01-" in date:
        break
    if "-02-" not in date:
        continue

    print(date)
    try:
        author = soup1.find("div",class_="article__author").find("a").text
    except:
        continue
    
    
    content = ""
    txt = soup1.find("div",class_="article__markdown publication-markdown article-body-print js-article-body js-article-body-print")
    txts = txt.find_all("p")
    for p in txts:
        text = p.text
        content += text+" "
        
        
    content = content.strip().replace("\n","")
    content = content.replace("\xa0"," ")
    
    f.write(title + "^" + author + "^" + date + "^" + url + "^" + content + "\n")
    
    
f.close()
```
